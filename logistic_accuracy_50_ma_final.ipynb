```python
 # Replace 'your_dataset.csv' with your dataset file

# Split data into features (X) and target variables (y_relevance and y_reliability)
X = data[['errors', 'accuracy', 'security_vulnerabilities','Algorithm','Model Used']]
y_relevance = data['Relevance']
y_reliability = data['reliability']

# Feature scaling (not always required for logistic regression, but it can improve performance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets for relevance
X_train_rel, X_test_rel, y_train_rel, y_test_rel = train_test_split(X_scaled, y_relevance, test_size=0.2, random_state=42)

# Split data into training and testing sets for reliability
X_train_rel, X_test_rel, y_train_rel, y_test_rel = train_test_split(X_scaled, y_reliability, test_size=0.2, random_state=42)

# Train Logistic Regression models for relevance and reliability
lr_model_rel = LogisticRegression()
lr_model_rel.fit(X_train_rel, y_train_rel)

lr_model_rel = LogisticRegression()
lr_model_rel.fit(X_train_rel, y_train_rel)

# Predict classes for relevance
y_pred_rel = lr_model_rel.predict(X_test_rel)

# Predict classes for reliability
y_pred_rel = lr_model_rel.predict(X_test_rel)

# Model evaluation for relevance
accuracy_rel = accuracy_score(y_test_rel, y_pred_rel)
print("Accuracy (Relevance):", accuracy_rel)
print("\nClassification Report (Relevance):")
print(classification_report(y_test_rel, y_pred_rel))

# Confusion matrix for relevance
conf_matrix_rel = confusion_matrix(y_test_rel, y_pred_rel)
print("\nConfusion Matrix (Relevance):")
print(conf_matrix_rel)

# Model evaluation for reliability
accuracy_rel = accuracy_score(y_test_rel, y_pred_rel)
print("Accuracy (reliability):", accuracy_rel)
print("\nClassification Report (reliability):")
print(classification_report(y_test_rel, y_pred_rel))

# Confusion matrix for reliability
conf_matrix_rel = confusion_matrix(y_test_rel, y_pred_rel)
print("\nConfusion Matrix (reliability):")
print(conf_matrix_rel)
```
